```python
from pyspark import SparkContext, SparkConf
conf = SparkConf().setAppName("Sample App").setMaster("local")
sc = SparkContext(conf=conf)
```

```python
data = [1, 2, 3, 4, 5]
distData = sc.parallelize(data)
```

```python
distData.reduce(lambda a, b: a + b)
```
